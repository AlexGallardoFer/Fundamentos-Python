```python
#######################################################################
# EJERCICIO 1A
####################################################################### 
#
# Escribe una función para calcular el consumo medio de combustible 
# de un vehículo.
#
# Entrada de datos:
#   - Kilómetros recorridos
#   - Litros de combustible consumidos
# 
# Salida de datos:
#   - Litros consumidos por kilómetro
#
#######################################################################

def ConsumoMedio(km, l):
    return l/km*100

try:
    while True:
        kilometros = input("Introduce la distancia recorrida en km: ")
        if kilometros.isdigit():
            break
    while True:
        litrosConsumidos = input("Introduce los litros consumidos: ")
        if litrosConsumidos.isdigit():
            break

    print(f"Se ha consumido una media de {ConsumoMedio(int(kilometros), int(litrosConsumidos)):1.2f} litros por 100 kilómetros.")

except Exception as err:
    print(err)
```

```python
#######################################################################
# EJERCICIO 2A
####################################################################### 
#
# Escribir funciones para transformar grados Celsius en Fahrenheit y
# viceversa.
#
# Fórmulas:
#   - Celsius = (5 / 9) * (Farhenheit - 32)
#   - Fahrenheit = (Celsius * (9 / 5)) + 32
#
# Entrada de datos:
#   - Grados Celsius/Fahrenheit
# 
# Salida de datos:
#   - Grados transformados
#
#######################################################################

def CelToFah(c):
    return (c*(9/5))+32

def FahToCel(f):
    return (5/9)*(f-32)

try:
    while True:
        celsius = input("Introduce unos grados Celsius: ")
        if celsius.isdigit():
            break
    while True:
        fahrenheit = input("Introduce unos grados Fahrenheit: ")
        if fahrenheit.isdigit():
            break

    print(f"{celsius}ºC son {CelToFah(float(celsius)):1.2f}ºF.\n")
    print(f"{fahrenheit}ºF son {FahToCel(float(fahrenheit)):1.2f}ºC.")

except Exception as err:
    print(err)
```
